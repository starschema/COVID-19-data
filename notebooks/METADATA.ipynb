## Create Snowflake schema documentation SQL Script


```python
def create_comment(row):
    if pd.isnull(row.column):
        return f"alter table {row.table} SET COMMENT='{row.description}';" 
    else:
        return f"comment on column {row.table}.{row.column} is '{row.description}';" 

ret = metadata.apply(create_comment, axis=1)

ret.to_csv( output_folder + "METADATA_COMMENT_DB_OBJECTS.sql",quoting=csv.QUOTE_NONE, sep="", doublequote=False, header=False, index=False)
```
